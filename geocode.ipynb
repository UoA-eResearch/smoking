```python
import googlemaps
import os
import pandas as pd
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
from tqdm import tqdm
tqdm.pandas()
```

```python
gmaps = googlemaps.Client(key=os.environ["GOOGLE_API_KEY"])
print(gmaps)
df = pd.read_excel("CPOHistory-UoA.xlsx", sheet_name="Full list - deprivation").dropna(subset="Address")
display(df)
```

```python
df["full_address"] = df.Premises + ", " + df.Address + ", " + df.Locality + ", Auckland"
df["full_address"]
```

```python
gmaps.geocode(df.full_address.iloc[0])
```

```python
df["geo"] = df.full_address.progress_apply(lambda address: gmaps.geocode(address, components={"country": "NZ"}))
```

```python
df["lat"] = df.geo.apply(lambda geo: geo[0]["geometry"]["location"]["lat"])
df["lng"] = df.geo.apply(lambda geo: geo[0]["geometry"]["location"]["lng"])
df
```

```python
df.to_csv("geocoded.csv", index=False)
```
